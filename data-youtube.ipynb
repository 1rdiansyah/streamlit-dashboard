```python
# Input data
import pandas as pd

data = pd.read_csv("data/trending.csv")
data.head()
```

```python
# Tampilkan kolom
data.columns
```

```python
# Buat data baru
data [['title', 'channel_name', 'category_id', 'view', 'like', 'dislike', 'comment', 'trending_time']]
```

```python
# Ubah 'category_id' jadi 'category_name'
import json

with open("data/category.json", "rb") as f:
    categories = json.load(f)
```

```python
categories = pd.json_normalize(categories, record_path="items")
categories_dict = {idx:title for idx, title in zip(categories['id'], categories['snippet.title'])}
```

```python
data['category'] = [categories_dict[str(x)] if str(x) in list(categories_dict.keys()) else "Unknown" for x in data['category_id']]
```

```python
data
```

```python
data['category'].value_counts()
```

```python
# Cek Null

data.isnull().sum()
```

```python
# Mengubah data null ke 0

data = data.fillna(0)
```

```python
# Mencari nilai trending date dan menghapus trending time

trending = pd.to_datetime(data['trending_time'])
data['trending_date'] = trending.dt.date
data = data.drop(['trending_time'], axis=1)
```

```python
data
```

```python
# Melihat data duplikat

data[['title', 'channel_name']].value_counts().reset_index(name='counts')
```

```python
used_data = data.drop_duplicates(['title', 'channel_name'], keep='first')
```

```python
used_data = used_data.reset_index(drop=True)
```

```python
used_data = used_data.drop(['category_id'], axis=1)
used_data.to_pickle('data/used_data.pickle')
```
